```python
import random
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageFilter
```

```python
img=cv2.imread('3.jpg')
unchanged_img=img
for_sharp=img
```

```python
fig, ax = plt.subplots()

ax.imshow(img)
fig.set_size_inches(7,7)
plt.show()
```

```python
row , col ,levels= img.shape
number_of_pixels = random.randint(300, 10000)
for i in range(number_of_pixels):

  # Pick a random y coordinate
  y_coord=random.randint(0, row - 1)

  # Pick a random x coordinate
  x_coord=random.randint(0, col - 1)

  # Color that pixel to white
  img[y_coord][x_coord] = 255
```

```python
number_of_pixels = random.randint(300, 10000)
for i in range(number_of_pixels):

  # Pick a random y coordinate
  y_coord=random.randint(0, row - 1)

  # Pick a random x coordinate
  x_coord=random.randint(0, col - 1)

  # Color that pixel to white
  img[y_coord][x_coord] = 0
```

```python
fig2, ax2 = plt.subplots()

ax2.imshow(img)
fig2.set_size_inches(7,7)
plt.show()
```

```python
#Averaging Filter
blur = cv2.blur(img,(5,5))
```

```python
fig3, ax3 = plt.subplots()

ax3.imshow(blur)
fig3.set_size_inches(7,7)
plt.show()
```

```python
#Median Filter
median = cv2.medianBlur(img,5)
```

```python
fig4, ax4 = plt.subplots()

ax4.imshow(median)
fig4.set_size_inches(7,7)
plt.show()
```

```python
#Rank Filter
img2=Image.fromarray(img)
rank = img2.filter(ImageFilter.RankFilter(size = 3, rank = 3))
```

```python
fig5, ax5 = plt.subplots()

ax5.imshow(rank)
fig5.set_size_inches(7,7)
plt.show()
```

```python
blurred = cv2.blur(unchanged_img,(10,10))

fig6, ax6 = plt.subplots()

ax6.imshow(blurred)
fig6.set_size_inches(7,7)
plt.show()
```

```python
#Sharpening Blurred image
#
#Perwitt Filter
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]], dtype=int)
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]], dtype=int)

x = cv2.filter2D(for_sharp, cv2.CV_16S, kernelx)
y = cv2.filter2D(for_sharp, cv2.CV_16S, kernely)

absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
Prewitt = cv2.addWeighted(absX,0.5,absY,0.5,0)
```

```python
fig7, ax7 = plt.subplots()

ax7.imshow(Prewitt)
fig7.set_size_inches(7,7)
plt.show()
```

```python
#Sobel Filter
kernelx = np.array([[1,2,1],[0,0,0],[-1,-2,-1]], dtype=int)
kernely = np.array([[-1,0,1],[-2,0,2],[-1,0,1]], dtype=int)

x = cv2.filter2D(for_sharp, cv2.CV_16S, kernelx)
y = cv2.filter2D(for_sharp, cv2.CV_16S, kernely)

absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
Sobel = cv2.addWeighted(absX,0.5,absY,0.5,0)
```

```python
fig8, ax8 = plt.subplots()

ax8.imshow(Sobel)
fig8.set_size_inches(7,7)
plt.show()
```

```python
#Laplace Filter
laplce=cv2.Laplacian(for_sharp,ddepth=-1,ksize=5)
```

```python
fig9, ax9 = plt.subplots()

ax9.imshow(laplce)
fig9.set_size_inches(7,7)
plt.show()
```

```python
kernel1 = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0,-1, 0]])

laplace2=cv2.filter2D(for_sharp,ddepth=-1,kernel=kernel1)
```

```python
fig9, ax9 = plt.subplots()

ax9.imshow(laplace2)
fig9.set_size_inches(7,7)
plt.show()
```
