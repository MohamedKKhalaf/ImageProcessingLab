# **Task 1**

## **Plot Image Histogram**

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np


# reads an input image
original = cv2.imread('/content/sample_data/img.jpg', 0)


# plot histogram with 255 bins
plt.hist(original.ravel(),256,[0,256])
plt.title("Image Histogram")
plt.xlabel("greyscale")
plt.ylabel("freq")
plt.show()
```

## **Plot Image Cumulative Histogram**

```python
# plot cumulative histogram
plt.hist(original.ravel(),256,[0,256], cumulative = True)
plt.title("Image Cumulative Histogram")
plt.xlabel("greyscale")
plt.ylabel("freq")
plt.show()
```

## **Perform Histogram Equalization**

```python
# creating a Histograms Equalization
# of a image using cv2.equalizeHist()
equ = cv2.equalizeHist(original)

# stacking images side-by-side
res = np.hstack((original, equ))

# show image input vs output
plt.imshow(res)
```

## **Plot Image Histogram after Equalization**

```python
# calculate mean value from RGB channels and flatten to 1D array
plt.hist(equ.ravel(),256,[0,256])
plt.title("Image Histogram")
plt.xlabel("greyscale")
plt.ylabel("freq")
plt.show()
```

## **Plot Image Cumulative Histogram after Equalization**

```python
# plot cumulative histogram
plt.hist(equ.ravel(),256,[0,256], cumulative = True)
plt.title("Image Cumulative Histogram")
plt.xlabel("greyscale")
plt.ylabel("freq")
plt.show()
```

## **Plot Image Cumulative Histogram**

```python
plt.hist(original.ravel(),256,[0,256], cumulative = True)
plt.title("Image Cumulative Histogram")
plt.xlabel("greyscale")
plt.ylabel("freq")
plt.show()
```

## **Apply contrast stertching on image pixels**

```python
norm_img1 = cv2.normalize(original, None, alpha=0, beta=1.2, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

norm_img1 = np.clip(norm_img1, 0, 1)
norm_img1 = (255*norm_img1).astype(np.uint8)

# stacking images side-by-side
res = np.hstack((original, norm_img1))

# show image input vs output
plt.imshow(res)
```
